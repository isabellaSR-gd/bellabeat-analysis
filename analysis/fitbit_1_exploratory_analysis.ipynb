<h4><center>About Dataset</h4></center>
This dataset generated by respondents to a distributed survey via Amazon Mechanical Turk between 03.12.2016-05.12.2016. Thirty eligible Fitbit users consented to the submission of personal tracker data, including minute-level output for physical activity, heart rate, and sleep monitoring. Individual reports can be parsed by export session ID (column A) or timestamp (column B). Variation between output represents use of different types of Fitbit trackers and individual tracking behaviors / preferences.

### Tentar com total distance ao invés de calories

```python
mean_distance = d_activity['TotalDistance'].mean() #4.6km
```

```python
mean_distance_id = pd.pivot_table(d_activity, values = 'TotalDistance', index= 'Id', aggfunc= 'mean').round(1)
```

```python
def avaliar_dist(value):
    if value > mean_distance:
        return "High average distance"
    else:
        return "Low average distance"
```

```python
weightlog_BMI = weightlog[['Id' ,'BMI_describe']]
weightlog_BMI
```

```python
distance_BMI = mean_distance_id.merge(weightlog_BMI, on='Id', how = 'left')
distance_BMI
```

```python
distance_BMI['DistanceEval'] = distance_BMI['TotalDistance'].apply(avaliar_dist)
distance_BMI.drop_duplicates()
```

```python
pd.crosstab(index= distance_BMI['BMI_describe'], columns= distance_BMI['DistanceEval'])
```
