Thirty eligible Fitbit users consented to the submission of personal tracker data, including minute-level output for physical activity, heart rate, and sleep monitoring. ### Checagem e Contagem de null steps por id

```python
# existe 14 ids com steps zero, mas todas as suas colunas nao estão com zero, saber qual contéudo das colunas dos ids com steps zro
id_a_verificar = [
    1844505072, 2320127002, 2347167796, 2873212765, 2891001357,
    4057192912, 4319703577, 4388161847, 4702921684, 6117666160,
    6290855005, 6391747486, 6775888955, 8253242879
]
for ids in id_a_verificar:
    print(f"Verificando se o ID {ids}")
    colunas_id = d_activity[d_activity['Id'] == ids].drop('Id', axis=1)
    
    # Check if all values in all columns (except 'Id') are zero
    todas_zeros = (colunas_id == 0).all().all()
    print("todas as suas colunas tem valor zero", todas_zeros)
```

```python
# contagem de null steps por id, sabendo que apenas um desses Ids (4388161847) não tem nenhum input de passos em seus registros
contagem_null_steps_id = d_activity.loc[d_activity['TotalSteps']== 0,'Id'].value_counts()
contagem_null_steps_id.sort_values(ascending=False)
```

Os dados com 0 inputs de passo podem representar pessoas que usam o fitbit apenas em dias de atividade física dedicada (como é inviável usar o fitbit e não registrar passos)

```python
# Ver contéudo das colunas desse ID 1844505072
d_activity[d_activity['Id']== 1844505072]
```

```python
# Quantas vezes o step zero aparece 
df_filtrado = d_activity[d_activity['TotalSteps'] == 0]
len(df_filtrado['TotalSteps'])
# Existe 61 registros com zero passos, divididos entre 14 ids, sendo que apenas uma delas não tem nenhum registro de passos
```

### Acrescentando infomação na Análise do  d_activity
* Do total de 427 registros, 61 deles são nulos (steps, atividade e minutos ativos zerados), divididos entre 14 ids.
* Esses 14 ids tem 1 ou mais dias com 0 registro de passos e atividade (sendo que apenas 1 id tem todos os seus 8 registros nulos)
