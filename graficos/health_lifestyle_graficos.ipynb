```python
# grafico apresenta a media de exercicios por grupo 
healh_mulheres = healh[healh['Gender'] == 'Female'] 
healh_bmi_cat = healh_mulheres['BMI_Category']
healh_idade_grupo = healh_mulheres['Age_Group']
ax = sns.barplot(healh_mulheres, x="Age_Group", y="Exercise_Hours_per_Week", estimator="mean", errorbar=None)
ax.bar_label(ax.containers[0], fontsize=10);
```
